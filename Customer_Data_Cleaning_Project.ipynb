Clean the Dataset

```python
# Clean 'Gender' column
df['Gender'] = df['Gender'].str.strip().str.capitalize()

# Clean 'Profession' column (handle missing values first, then format)
df['Profession'] = df['Profession'].fillna('Unknown').str.strip().str.title()
```

```python
# Rename columns to snake_case format
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('$', '')

# Let's see the new column names
print("New Column Names:", df.columns.tolist())
```

```python
df = df.rename(columns={
    'annual_income_': 'annual_income',
    'spending_score_1-100': 'spending_score'
})
```

```python
# Remove or cap impossible numeric values
invalid_experience = df[df['work_experience'] > df['age']]
print(f"Rows where experience exceeds age: {len(invalid_experience)}")

# Replace missing values in numeric columns
df['work_experience'] = df['work_experience'].fillna(0).astype(int)
df['family_size'] = df['family_size'].fillna(1).astype(int)
```

```python
# Strip out any rows that have completely identical profiles
df = df.drop_duplicates(subset=['customerid'], keep='first')
```

Final Cleaned Dataset

```python
print("--- Final Cleaned Dataset ---")
print(df.head())
```

```python
# Save and download the file from Colab
cleaned_filename = 'Cleaned_Customers.csv'
df.to_csv(cleaned_filename, index=False)
```
